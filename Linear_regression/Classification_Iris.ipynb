```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn import neighbors
```

```python
# Look at what sklearn gives us
iris = datasets.load_iris()
iris.keys()
```

```python
# Read in and structure the data nicely
irData = pd.DataFrame(iris['data'], columns=iris['feature_names'])
# Adding the species classifications to the DF
identNum = pd.Series(iris['target'])
identName = pd.Series([iris['target_names'][i] for i in identNum])
colors = sns.color_palette("tab10")
colorName = pd.Series([colors[i] for i in identNum])
irData['ident'] = identNum
irData['ident2'] = identName
irData['color'] = colorName
# For plotting in the cor-scat-plot below
featureNames = iris['feature_names']
nfeatures = len(iris['feature_names'])

irData.head(3)
```

```python
# This works though, but takes away a little of my control
panelWidth=3
panelHeight=3
marginWidth = 0.5
marginHeight = 0.5
figureWidth=(panelWidth * nfeatures) + (marginWidth * (nfeatures+1))
figureHeight=(panelHeight * nfeatures) + (marginHeight * (nfeatures+1))

# Using subplots and buffering some space
fig, axs = plt.subplots(nrows=nfeatures, ncols=nfeatures, figsize=(figureWidth,figureHeight))
plt.subplots_adjust(hspace=0.3, wspace=0.3)
# Loop over my features and plot them against each other.
for x,featureX in enumerate(featureNames):
    for y,featureY in enumerate(featureNames):
        axs[x,y].scatter(irData[featureX], irData[featureY], color=irData['color'])
        axs[x,y].set_xlabel(featureX)
        axs[x,y].set_ylabel(featureY)
# Only the diagonal and up actually matters, the other ones are repeats.
plt.show()
```

I can see how the variables interact with each other up above, but now I am curious if just from one of my features I can already classify my 3 outcome variables. From my corr-scat-plot, it looks like petal width and petal length will let me classify the blue colored flowers.

# Boxplots

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10,10))
plt.subplots_adjust(hspace=0.5, wspace=0.5)
for row in range(2):
    for column in range(2):
        sns.boxplot(y=featureNames[2*(row)+column], x="ident2", data=irData, ax=axs[row,column])
        ax = sns.swarmplot(y=featureNames[2*(row)+column], x="ident2", data=irData, ax=axs[row,column], s=2, color='black')
        ax.set_xticklabels(ax.get_xticklabels(),rotation = 30)
```

Just as we can see from the corr-scat-plot, petal length and width are perfect predictors of the Sentosa plant. I'm curious what a PCA of this would look like! Interestingly enough this actually just looks a lot like petal length vs sepal width corr-scat-plot up above. 

```python
pca = PCA(n_components=2)
pca.fit_transform(irData.iloc[:,0:4])
print(np.append(featureNames, pca.components_))
lowerD = pd.DataFrame(pca.fit_transform(irData.iloc[:,0:4]))
lowerD.plot(kind='scatter', x=0, y=1, c=irData['color'])
```
